```python
import jpybinding as jpb
import matplotlib.pyplot as plt
import numpy as np
import scipy as sci
```

```python
np.array([[-2],[-3]]).shape
```

```python
lat=jpb.Lattice(a1=[0.1,0],a2=[0,0.1])

# lat.add_sublattices(
#     ("A", [0,0],np.eye(2)),
#     ("B", [0.5,0.5])
# )

# lat.add_hoppings(
#     ([1,0], "A", "B", np.array([[-2],[-3]])),
# )

lat.add_sublattices(
    ("A", [0,0])
)

lat.add_hoppings(
    ([1,0], "A", "A", -1),
    ([0,1], "A", "A", -1),
)



lat.plot()
```

```python
lat.sub_index_orb
lat.norbs
```

```python
model=jpb.Model(lat,'k',16,16)
```

```python
Gamma=[0,0]
X=[10*np.pi,10*np.pi]
k_points=[Gamma,X,Gamma]


```

```python
solver=jpb.Solver(model)
```

```python
ene_mat=solver.calc_bands(k_points)
```
